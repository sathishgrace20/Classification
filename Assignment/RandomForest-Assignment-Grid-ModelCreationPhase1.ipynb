```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
dataset["classification_yes"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid={'criterion':['gini', 'entropy', 'log_loss'],'max_features':['sqrt', 'log2', None]}
classifier=GridSearchCV(RandomForestClassifier(),param_grid, n_jobs=-1,refit=True, verbose=3,
    return_train_score=True)       
```

```python
classifier=classifier.fit(X_train,y_train)
```

```python
classifier
```

```python
print(classifier.best_params_)
```

```python
print(classifier.cv_results_)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')
print("The f1_macro value for best parameter {}:".format(classifier.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,classifier.predict_proba(X_test)[:,1])
```
